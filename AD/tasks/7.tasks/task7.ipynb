### Задача 1.

Перед обучением линейной регрессии вы могли совершать некоторые преобразования над признаками. Какие преобразования стоит совершить перед обучением решающего дерева?

В частности, ответьте на вопрос, имеет ли смысл примененять следующие преобразования:
* монотонные функции от одного признака;
> Ответ: Нет, смысла не имеет, т.к. множество возможных разбиений вершины $X$ на $X_l$ и $X_r$ по данному признаку не поменяется. 
* немонотонные функции от одного признака;
> Ответ: возможно имеет смысл, например $|x - t_0|$ в случае с температурой, ведь слишком высокая и низкая температура одновременно неприемлимы. Почему возможно? Потому что потенциально дерево сможет само совершить несколько соответствующих разделений эквивалентных одному разделению измененённого признака, а может и не совершить, опираясь на свою жадность при совершении первого "не слишком выгодного" разделения. Поэтому вероятно такие преобразования имеют смысл (особенно при малой глубине дерева), но подбирать параметры преобразований - тоже будет муторно(гиперпараметр получается).
* линейные функции от нескольких признаков;
> Ответ: аналогично предыдущему ответу: при малой глубине такие гипотезы (что лин. комб. лучше чемдва признака по отдельности) могут и оправдаться. Например признак А - потрачено времени на странице личного кабинета, признак B - потрачено времени на странице с товарами, target - купил ли пользователь что-нибудь за эту сессию использования приложения. Ну и в целом логичное предположение сделать признак А+B и построить по нему дерево глубины 1(допустим такое ограничение). Однако при большей глубине признаки А и B сами решат как им лучше и комбинация признаков не будет иметь смысла.
* нелинейные функции от нескольких признаков;
> Ответ: аналогично предыдущему ответу; Добавим лишь, что подбор параметров таких комбинаций признаков займёт также немалое время, возможно в этом контексте это будет не совсем целесообразно.
* стандартизация признаков;
> Ответ: стандартизацию проводить нет смысла, ведь она не меняет порядок значений(монотонная функция), в рассмотренных нами StandardScaler и MinMaxScaler.
* преобразования категориальных признаков.
> Ответ: при работе с категориальными признаками в решающих есть несколько подходов: построение более двух(например $N$) сыновей из вершины(по количеству категорий $= N$) или сопоставление одному категориальному признаку с $N$ категориями $2^N$ бинарных признаков вида $F_i(var) = var \in M_i$, где $M_i$ - очередное подмножество категорий. Также можно переупорядочить категории(сделать эту функцию более "гладкой"), чтобы операция $<$ имела больше смысла. Делать OneHotEncoding также возможно, однако чтобы признаки с большим количеством категорий влияли на предсказание - нужна глубина побольше.

### Задача 2.

Рассмотрим данные из <a href="https://mipt-stats.gitlab.io/courses/ad_fivt/task6.html" target="_blank">предыдущего домашнего задания</a>. Скачайте данные и разделите их на три части &mdash; обучающую, валидационную и тестовую в соотношении 3:1:1.

```python
data = pd.read_csv('bikes.csv')
data = data.reindex(columns=['season', 'mnth', 'holiday', 'weekday', 'workingday', 
                             'weathersit', 'temp', 'atemp', 'hum', 'windspeed', 'cnt'])

train_ratio = 0.6
val_ratio = 0.2
test_ratio = 0.2

train, test = train_test_split(data, test_size=test_ratio)
train, val = train_test_split(train, test_size=val_ratio/(train_ratio + val_ratio)) 

train.shape, val.shape, test.shape
```

```python
train.shape, val.shape, test.shape
```

Выполните необходимые преобразования признаков

```python
def drop_outliers(data, real_features):
    return data[(np.abs(scipy.stats.zscore(data[real_features])) < 3).all(axis=1)]

categorial_features = ['season', 'mnth', 'holiday', 'weekday', 'weathersit']
real_features = ['temp', 'atemp', 'hum', 'windspeed']
target_feature = 'cnt'

def prepare(data, encoder):
    d = copy.deepcopy(data)
    d = drop_outliers(d, real_features)
    train_cat = encoder.fit_transform(d[categorial_features])
    train = np.hstack([d[real_features], train_cat])
    cat_labels = [[title+", "+str(x) for x in l] for l, title in zip(encoder.categories_, encoder.feature_names_in_)]
    labels = copy.deepcopy(real_features)
    for x in cat_labels:
        labels += x
    return train, np.array(d[target_feature]), labels

encoder = OneHotEncoder(sparse=False)
input_train, target_train, labels = prepare(train, encoder)
input_val, target_val, labels = prepare(val, encoder)
input_test, target_test, labels = prepare(test, encoder)
len(labels), input_test.shape[1]
```

Обучите решающее дерево глубины 3 и визуализируйте его

```python
model = DecisionTreeRegressor(max_depth=3, criterion='absolute_error')
model.fit(input_train, target_train)
```

Визуализируем

```python
def get_graphviz_graph(decision_tree_regressor, feature_names):
    dot_data = export_graphviz(
        decision_tree_regressor, out_file=None, 
        feature_names=feature_names, class_names='Price',  
        filled=True, rounded=True, special_characters=True
    )
    return graphviz.Source(dot_data)

graph = get_graphviz_graph(model, labels)
graph
```

Подберите оптимальное значение максимальной глубины дерева. Для этого обучите решающие деревья с выставленным значением соответствующего гиперпараметра и посчитайте точность предсказаний каждого из них на валидационной выборке. Постройте соответствующий график.

```python
depth_range = np.arange(1, 30)
models = []
for depth in depth_range:
    model = DecisionTreeRegressor(max_depth=depth, criterion='absolute_error')
    model.fit(input_train, target_train)
    models += [model]
```

Визуализируем

```python
sns.set(font_scale=1.4, palette='summer')
colors = ['red', 'blue']
index = 0

def mean_absolute_percentage_error(y_true, y_pred):
    return 100 * (np.abs((y_true - y_pred) / y_true)).mean()

def get_error_rates(models, input_, target_):
    rmse = []
    mae = []
    mape = []
    for model in models:
        preds_ = model.predict(input_)
        rmse.append(metrics.mean_squared_error(target_, preds_) ** 0.5)
        mae.append(metrics.mean_absolute_error(target_, preds_))
        mape.append(mean_absolute_percentage_error(target_, preds_))
    return rmse, mae, mape

def plot_error_rates(rmse, mae, mape, arg_range, legend, xlabel='Варьируемый параметр'):
    global index
    titles = ["RMSE", "MAE", "MAPE"]
    arrays = [rmse, mae, mape]
    for i in range(len(titles)):
        ax = plt.subplot(1, 3, i + 1)
        ax.plot(arg_range, arrays[i], label=legend, color=colors[index], alpha=0.5)
        ax.set_title(titles[i])
        ax.legend(loc=1)
        ax.set_xlabel(xlabel)
        ax.set_ylabel("Значение ошибки")
    index = (index + 1) % 2

fig = plt.figure(figsize=(21, 7))
# plot_error_rates(*get_error_rates(models, input_train, target_train), depth_range, "Трениров.", "Глубина дерева")
plot_error_rates(*get_error_rates(models, input_val, target_val), depth_range, "Валид.", "Глубина дерева") 
```

Вывод: оптимальная **глубина = 3**.

Аналогичным образом подберите минимальное число объектов в листе, зафиксировав при этом подобранное значение максимальной глубины.

```python
leaf_range = np.arange(1, 100)
models = []
for min_samples_leaf in leaf_range:
    model = DecisionTreeRegressor(max_depth=3, min_samples_leaf=min_samples_leaf, criterion='absolute_error')
    model.fit(input_train, target_train)
    models += [model]
```

Визуализируем

```python
fig = plt.figure(figsize=(21, 7))
plot_error_rates(*get_error_rates(models, input_val, target_val), leaf_range, "Валид.", "Сэмплов в листе")
```

Минимальное число **сэмплов в листе = 19**

Визуализируйте дерево при оптимальных гиперпараметрах и посчитайте его точность на тестовой части выборки.

```python
model = DecisionTreeRegressor(max_depth=3, min_samples_leaf=19, criterion='absolute_error')
model.fit(input_train, target_train)

errors = [error[0] for error in [*get_error_rates([model], input_test, target_test)]]
print(f"RMSE: {errors[0]}\nMAE: {errors[1]}\nMAPE: {errors[2]}")

graph = get_graphviz_graph(model, labels)
graph
```

Решающее дерево позволяет оценить *важность признаков* &mdash; некоторая величина, зависящая от того
* насколько часто тот или иной признак использовался при разбиении листа, 
* и насколько сильно при этом изменился критерий информативности.

У обученной модели значения важности признаков лежат в поле `feature_importances_`. Визуализируйте их с помощью `seaborn.barplot`.

```python
plt.figure(figsize=(9, 9))
with sns.axes_style('whitegrid'):
    sns.barplot(y=labels, x=model.feature_importances_, orient='h')
```

Не преобразовывая категориальные признаки получим:

```python
def prepare_no_onehot(data):
    d = copy.deepcopy(data)
    d = drop_outliers(d, real_features)
    train = np.hstack([d[real_features], d[categorial_features]])
    labels = real_features + categorial_features
    return train, np.array(d[target_feature]), labels

input_train, target_train, labels = prepare_no_onehot(train)
input_val, target_val, labels = prepare_no_onehot(val)
input_test, target_test, labels = prepare_no_onehot(test)
len(labels), input_test.shape[1]
```
