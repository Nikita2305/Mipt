С помощью кода ниже загрузите встроенные в `sklearn` <a target="_blank" href="https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_20newsgroups.html#sklearn.datasets.fetch_20newsgroups">данные</a>, в которых представлены различные новостные сюжеты, разделенные на 20 тематических групп.

```python
# удаляем заголовки и подписи
remove = ('headers', 'footers')
# зафиксируем зерно случайности
random_state = 42

# Извлекаем обучающую и тестовую части перемешивая случайным образом
data = fetch_20newsgroups(
    subset='all', shuffle=True, random_state=random_state, remove=remove
)

core = dict()
core['target'] = data['target']
core['data'] = data['data']
    
df = pd.DataFrame(core)
df.head()
```

Выделим валидационную

```python
train_ratio = 0.6
val_ratio = 0.2
test_ratio = 0.2

train_val, test = train_test_split(df, test_size=test_ratio)
train, val = train_test_split(train_val, test_size=val_ratio/(train_ratio + val_ratio)) 

train.shape, val.shape, test.shape
```

С помощью методов, изученных в рамках нашего курса, постройте классификатор новостных сюжетов на тематические группы. При построении классификатора **запрещено пользоваться тестовой выборкой**. Для настройки классификатора, например, выбора оптимальных гиперпараметров, можно разделить обучающую выборку на две части &mdash; непосредственно обучающую и валидационную. Первую часть использовать для непосредственного обучения, а вторую &mdash; для настройки, пытаясь по ней оптимизировать качество.

```python
def get_model(**kwargs):
    vectorizer = CountVectorizer(**kwargs)
    vectorizer.fit(train_val['data'])
    train_X, train_Y = vectorizer.transform(train['data']).toarray(), train['target']
    bayes = MultinomialNB()
    bayes.fit(train_X, train_Y)
    return vectorizer, bayes

def get_accuracy(data):
    ret = np.zeros(len(data))
    print(f"Total: {len(data)}")
    for i, arr in enumerate(data):
        print(f"{i}: value = {arr}")
        if (arr[0] < arr[1]):
            vectorizer, model = get_model(min_df=arr[0], max_df=arr[1])
            val_X, val_Y = vectorizer.transform(val['data']).toarray(), val['target']
            np.put(ret, i, accuracy_score(val_Y, model.predict(val_X)))
    return ret
```

```python
grid = np.mgrid[0.01:0.1:0.01, 0.01:0.1:0.01]
_, h, w = grid.shape
grid_labels = get_accuracy(grid.reshape((2, h*w)).T).reshape(h, w)
print(grid_labels)
```

```python
plt.figure(figsize=(12, 12))
ax = plt.axes()
ax.pcolormesh(grid[0], grid[1], grid_labels[:, :],  shading='gouraud', cmap='Greens', alpha=1)
ax.set_title("accuracy_score")
ax.set_xlabel("min_df")
ax.set_ylabel("max_df");
```

Подберём $min\_df$ при фикисрованном $max\_df=0.05$

```python
mindf_range = np.arange(0.001, 0.02, 0.001)
accuracy = []
for mindf in mindf_range:
    print(f"mindf={mindf}")
    vectorizer, model = get_model(min_df=mindf, max_df=0.05)
    val_X, val_Y = vectorizer.transform(val['data']).toarray(), val['target']
    accuracy += [accuracy_score(val_Y, model.predict(val_X))]
```

```python
plt.figure(figsize=(12, 12))
ax = plt.axes()
ax.plot(mindf_range, accuracy);
ax.set_title("Подбор параметра $min\_df$")
ax.set_xlabel("$min\_df$")
ax.set_ylabel("accuracy score")
```

Вывод - выберем\
$min\_df = 0.002$\
$max\_df = 0.05$

Посчитайте качество на тестовой выборке.

```python
vectorizer, model = get_model(min_df=0.002, max_df=0.05)
test_X, test_Y = vectorizer.transform(test['data']).toarray(), test['target']
print(f"Accuracy: {accuracy_score(test_Y, model.predict(test_X))}")
```

### Задача 2.
Пусть $\mathscr{X} = \mathbb{R}^2$ — пространство признаков, $\mathscr{Y} = \{0, 1\}$ — множество классов. Рассматривается квадратичный дискриминантный анализ. Условное распределение $X$ при условии $Y=k$ равно $\mathcal{N}(a_k, \Sigma_k)$. Приведите примеры таких параметров $a_k, \Sigma_k$ и вероятностей $\mathsf{P}(Y = k)$, при которых разделяющая поверхность является
* гиперболой;
* параболой;
* двумя параллельными прямыми;
* двумя пересекающимися прямыми.

### Часть 1. Общие рассуждения.
Чтобы задать разделяющую поверхность в данном случае запишем:\
$\ln{\pi_1} - \frac{1}{2} \ln{\det{\Sigma_1}} - \frac{1}{2} (x-a)^T \Sigma_1^{-1} (x-a) = 
\ln{\pi_2} - \frac{1}{2} \ln{\det{\Sigma_2}} - \frac{1}{2} (x-b)^T \Sigma_2^{-1} (x-b)$

Положим $a = (0, 0)$, просто сдвинув все сэмплы на старый вектор $-a_1$.

$x^T \Sigma_1^{-1} x = (x-b)^T \Sigma_2^{-1} (x-b) +
2\ln{\pi_1} - \ln{\det{\Sigma_1}} - 2\ln{\pi_2} + \ln{\det{\Sigma_2}}$

Положим $
\Sigma_1^{-1} = \begin{pmatrix}
p & q\\
q & r
\end{pmatrix}$ и 
$\Sigma_2^{-1} = \begin{pmatrix}
f & g\\
g & h
\end{pmatrix}$

Тогда $\Sigma_1 = \frac{1}{pr - q^2} * \begin{pmatrix}
r & -q\\
-q & p
\end{pmatrix}$ и 
$\Sigma_2 = \frac{1}{fh - g^2} * \begin{pmatrix}
h & -g\\
-g & f
\end{pmatrix}$

Запишем урвнение ($x, y \in \mathbb{R}$):\
$px^2 + 2qxy + ry^2 = f(x-b_1)^2 + 2g(x-b_1)(y-b_2) + h(y-b_2)^2 + C_1$,\
где $С_1 = 2\ln{\pi_1} - \ln{\det{\Sigma_1}} - 2\ln{\pi_2} + \ln{\det{\Sigma_2}}$

$x^2(p - f) + xy(2q - 2g) + y^2(r - h) + x(2fb_1 + 2gb_2) + y(2hb_2 + 2gb_1) - fb_1^2 - 2gb_1b_2 - hb_2^2 - C_1 = 0$

Или $C_2 = -C_1 - fb_1^2 - 2gb_1b_2 - hb_2^2$, тогда\
$x^2(p - f) + 2xy(q - g) + y^2(r - h) + 2x(fb_1 + gb_2) + 2y(hb_2 + gb_1) + C_2 = 0$

Далее будем обращаться к последнему уравнению.

### Часть 2. Решение.

Для подобра соответствующих параметров будем использовать [эту статью на вики](https://ru.wikipedia.org/wiki/%D0%9A%D1%80%D0%B8%D0%B2%D0%B0%D1%8F_%D0%B2%D1%82%D0%BE%D1%80%D0%BE%D0%B3%D0%BE_%D0%BF%D0%BE%D1%80%D1%8F%D0%B4%D0%BA%D0%B0#%D0%9A%D0%BB%D0%B0%D1%81%D1%81%D0%B8%D1%84%D0%B8%D0%BA%D0%B0%D1%86%D0%B8%D1%8F_%D0%BA%D1%80%D0%B8%D0%B2%D1%8B%D1%85_%D0%B2%D1%82%D0%BE%D1%80%D0%BE%D0%B3%D0%BE_%D0%BF%D0%BE%D1%80%D1%8F%D0%B4%D0%BA%D0%B0_%D0%BE%D1%82%D0%BD%D0%BE%D1%81%D0%B8%D1%82%D0%B5%D0%BB%D1%8C%D0%BD%D0%BE_%D0%B7%D0%BD%D0%B0%D1%87%D0%B5%D0%BD%D0%B8%D0%B9_%D0%B8%D0%BD%D0%B2%D0%B0%D1%80%D0%B8%D0%B0%D0%BD%D1%82%D0%BE%D0%B2)

В ней используется обозначение нашего уравнения:

$a_{11} x^2 + 2 a_{12} xy + a_{22} y^2 + 2 a_{13} x + 2 a_{23} y + a_{33} = 0$

Чтобы это было релевантно - хотим, чтобы хотя бы одно из чисел $(p-f), (q-g), (r-h)$ было отлично от нуля. Иными словами матрицы ковариаций отличались.

### Гипербола и пересекающиеся прямые.

Для получения этих случаев **необходимо условие**: $(r-h)(p-f) - (q-g)^2 < 0$. Буковок целых шесть, мы уверены, что у нас это получится. Стоит не забывать о свойствах ковариации, в том числе определитель $\Sigma_k$ (а значит и $\Sigma_k^{-1}$) в данном случае должен быть положителен.\
Например $(r-h) = 1, (p-f) = 1, (q-g)=2$

Положим\
$r = 3, h = 2$\
$p = 3, f = 2$\
$q = 2, g = 0$, т.е.

$\Sigma_1 = \frac{1}{5} * \begin{pmatrix}
3 & -2\\
-2 & 3
\end{pmatrix}$ и 
$\Sigma_2 = \frac{1}{2} * \begin{pmatrix}
1 & 0\\
0 & 1
\end{pmatrix}$

После удовлетворения первого условия - разделяющим становится **второе условие**:

$\det
\begin{pmatrix}
a_{11} & a_{12} & a_{13}\\
a_{12} & a_{22} & a_{23}\\
a_{13} & a_{23} & a_{33}
\end{pmatrix} \stackrel{?}{=} 0$

Положим $b = (0, 0)$, тогда $a_{13} = a_{23} = 0$ - коэффициенты при $x$ и $y$ (см. уравнение в общих рассуждениях).
Тогда второе условие сводится к $a_{33} \stackrel{?}{=} 0$ - свободный член, который мы ранее обозначили $C_2$.
Пришло время его расписать и выбрать $\pi_k$ так, чтобы удовлетворить и не удовлетворить это условие.

$C_2 = -C_1 - fb_1^2 - 2gb_1b_2 - hb_2^2 = -C_1 - 0 - 0 - 0 = -C_1 \stackrel{?}{=} 0$

$C_1 = 2\ln{\pi_1} - \ln{\det{\Sigma_1}} - 2\ln{\pi_2} + \ln{\det{\Sigma_2}} = \ln(\frac{\pi_1^2}{\pi_2^2} * \frac{\det{\Sigma_2}}{\det{\Sigma_1}}) \stackrel{?}{=} 0$

Что эквивалентно: $\frac{\pi_1^2}{\pi_2^2} \stackrel{?}{=} \frac{\det{\Sigma_1}}{\det{\Sigma_2}} = \frac{4}{5}$

Чтобы получить пересекающиеся прямые, это условие должно выполниться, а значит $\pi_1 = \frac{2}{2 + \sqrt{5}}, \pi_2 = \frac{\sqrt{5}}{2 + \sqrt{5}}$.\
Чтобы получить гипепболу, выберем любые другие $\pi_k$. Например $\pi_k = \frac{1}{2}$

**Итого ответ**, в обоих случаях:

$\Sigma_1 = \frac{1}{5} * \begin{pmatrix}
3 & -2\\
-2 & 3
\end{pmatrix}$ и 
$\Sigma_2 = \frac{1}{2} * \begin{pmatrix}
1 & 0\\
0 & 1
\end{pmatrix}$

$a = b = (0, 0)$

В случае гиперболы $\pi_k = \frac{1}{2}$\
В случае пересекающихся прямых: $\pi_1 = \frac{2}{2 + \sqrt{5}}, \pi_2 = \frac{\sqrt{5}}{2 + \sqrt{5}}$


### Парабола и пара параллельных прямых

**Напомним:** $x^2(p - f) + 2xy(q - g) + y^2(r - h) + 2x(fb_1 + gb_2) + 2y(hb_2 + gb_1) + C_2 = 0$\
$a_{11} x^2 + 2 a_{12} xy + a_{22} y^2 + 2 a_{13} x + 2 a_{23} y + a_{33} = 0$

Для получения этих случаев **необходимо условие**: $(r-h)(p-f) - (q-g)^2 = 0$. Буковок целых шесть, мы уверены, что у нас это получится. Стоит не забывать о свойствах ковариации, в том числе определитель $\Sigma_k$ (а значит и $\Sigma_k^{-1}$) в данном случае должен быть положителен.\
Например $(r-h) = 1, (p-f) = 1, (q-g) = 1$

Положим\
$r = 2, h = 1$\
$p = 2, f = 1$\
$q = 1, g = 0$, т.е.

$\Sigma_1 = \frac{1}{3} * \begin{pmatrix}
2 & -1\\
-1 & 2
\end{pmatrix}$ и 
$\Sigma_2 = \begin{pmatrix}
1 & 0\\
0 & 1
\end{pmatrix}$

После удовлетворения первого условия - разделяющим становится **второе условие**:

$\det
\begin{pmatrix}
a_{11} & a_{12} & a_{13}\\
a_{12} & a_{22} & a_{23}\\
a_{13} & a_{23} & a_{33}
\end{pmatrix} \stackrel{?}{=} 0$

Положим $b = (1, 0)$

$\det
\begin{pmatrix}
a_{11} & a_{12} & a_{13}\\
a_{12} & a_{22} & a_{23}\\
a_{13} & a_{23} & a_{33}
\end{pmatrix} = \det
\begin{pmatrix}
1 & 1 & 1\\
1 & 1 & 0\\
1 & 0 & C_2
\end{pmatrix} = -1 \neq 0$

**Это был кейс параболы.** Становятся не важны все остальные параметры.

**Теперь параллельные прямые.**

Положим $b = (0, 0)$, тогда:

$\det
\begin{pmatrix}
a_{11} & a_{12} & a_{13}\\
a_{12} & a_{22} & a_{23}\\
a_{13} & a_{23} & a_{33}
\end{pmatrix} = \det
\begin{pmatrix}
1 & 1 & 0\\
1 & 1 & 0\\
0 & 0 & C_2
\end{pmatrix} = 0$

Помимо этого для параллельных прямых осталось одно условие: $a_{33}(a_{11} + a_{12}) - a_{13}^2 - a_{23}^2 < 0$\
$C_2 * (1 + 1) - 0 - 0 < 0$\
Что эквивалентно $C_2 < 0$\
$C_2 = -C_1 - fb_1^2 - 2gb_1b_2 - hb_2^2 = -C_1 - 0 - 0 - 0 = -C_1 < 0$\
$C_1 = 2\ln{\pi_1} - \ln{\det{\Sigma_1}} - 2\ln{\pi_2} + \ln{\det{\Sigma_2}} > 0$

$C_1 = \ln(\frac{\pi_1^2}{\pi_2^2} * \frac{\det{\Sigma_2}}{\det{\Sigma_1}}) >0$

$\frac{\pi_1^2}{\pi_2^2} * \frac{\det{\Sigma_2}}{\det{\Sigma_1}} > 1$

$\frac{\pi_1^2}{\pi_2^2} > \frac{\det{\Sigma_1}}{\det{\Sigma_2}} = \frac{1}{3}$

Например $\pi_k = \frac{1}{2}$

**Итого ответ**, в обоих случаях:

$\Sigma_1 = \frac{1}{3} * \begin{pmatrix}
2 & -1\\
-1 & 2
\end{pmatrix}$ и 
$\Sigma_2 = \begin{pmatrix}
1 & 0\\
0 & 1
\end{pmatrix}$

$a = (0, 0)$

В случае параболы $b = (1, 0); \pi_k = \frac{1}{2}$\
В случае параллельных прямых: $b = (0, 0), \pi_k = \frac{1}{2}$
