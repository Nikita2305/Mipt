### Задача 1.

Рассмотрим одномерную линейную регрессию без свободного параметра $y(x) = \theta x$, где $x\in \mathbb{R}$ и $\theta \in \mathbb{R}$. Предполагаем, что данные получены по правилу
$$Y_i = \theta x_i + \varepsilon_i,$$
где $i \in \{1, ..., n\}$, числа $x_i \in \mathbb{R}$ неслучайны, а $\varepsilon_i$ &mdash; случайные ошибки.

**1.** Посчитайте оценку $\widehat{\theta}$ методом наименьших квадратов.

**2.** Выпишите формулы градиентного спуска (GD) и стохастического градиентного спуска (SGD) для поиска $\widehat{\theta}$.

**3.** Пусть $\mathsf{E}\varepsilon_i = 0$, а $x \in \mathbb{R}$ &mdash; новый объект. Посчитайте $\mathsf{E}\widehat{\theta}$ и $\mathsf{E}\widehat{y}(x)$, где $\widehat{y}(x) = \widehat{\theta} x$.

**4.** Пусть $\mathsf{E}\varepsilon_i = 0$ и $\mathsf{D}\varepsilon_i = \sigma^2$, причем все $\varepsilon_i$ независимы, а $x \in \mathbb{R}$ &mdash; новый объект. Посчитайте $\mathsf{D}\widehat{\theta}$ и $\mathsf{D}\widehat{y}(x)$, где $\widehat{y}(x) = \widehat{\theta} x$.



*Замечания.* 
1. В ответе во всех пунктах могут оставаться суммы $n$ слагаемых, но не должно быть операций с векторами или матрицами.
2. Формулами мат. ожидания и дисперсии с занятия пользоваться нельзя, но стоит сверить ответ.
2. **Прочитай требования к формату решения в правилах сдачи задания.**

### Решение #1.

Рассмотрим $F(\theta) = \sum_{i=1}^n(Y_i - x_i\theta)^2$.
Решим задачу $F \rightarrow \min\limits_{{\theta} \in \mathbb{R}}$.\
$F(\theta) = \sum_{i=1}^n Y_i^2 - \theta*2\sum_{i=1}^n Y_i x_i +\theta^2 * \sum_{i=1}^n x_i^2$
 
$F'(\theta) = -2*\sum_{i=1}^n Y_i x_i + 2*\theta\sum_{i=1}^n x_i^2$\
$F'(\theta) = 0 \iff \theta = \frac{\sum_{i=1}^n Y_i x_i}{\sum_{i=1}^n x_i^2}$

Получаем оценку $\widehat{\theta} = \frac{\sum_{i=1}^n Y_i x_i}{\sum_{i=1}^n x_i^2}$

### Решение #2.

Вновь решаем задачу $F \rightarrow \min\limits_{{\theta} \in \mathbb{R}}$.
* **Градиентный спуск**\
Выберем $\theta_0$ - некоторое приближение искомой оценки $\widehat{\theta}$.\
Выберем также $\eta \in [10^{-4}, 10^{-1}]$.\
Рассмотрим последовательность $\theta_{t+1} = \theta_t - \eta*F'(\theta_t)$, где $F'(\theta) = -2*\sum_{i=1}^n Y_i x_i + 2*\theta\sum_{i=1}^n x_i^2$\
Устремляя $t\rightarrow\inf$, получим $\theta_t\rightarrow\widehat{\theta}$, где $\widehat{\theta}$ - искомая оценка.
* **Стохастический градиентный спуск. Mini batch версия.**\
В этом случае последовательность будет выглядеть так:\
$\theta_{t+1} = \theta_t - \eta*\sum_{j=1}^k (\theta_t x_{i_j}^2 - Y_{i_j} x_{i_j})$,\
где\
$\quad k$ - гиперпараметр, размер пакета.\
$\quad I = [i_1, ..., i_k], i_j \in \{1,2,...,n\}$, выбрано случайно для каждого j, t.

$\quad\quad$*Примечание: под $\eta$ занесли $\frac{2n}{k}$*

### Решение #3.

**Условие**\
$\widehat{\theta} = \frac{\sum_{i=1}^n Y_i x_i}{\sum_{i=1}^n x_i^2}$\
$\mathsf{E}\varepsilon_i = 0$

**Часть 1.**\
$\mathsf{E}\widehat{\theta} = \mathsf{E}\frac{\sum_{i=1}^n Y_i x_i}{\sum_{i=1}^n x_i^2} =
\mathsf{E}\frac{\sum_{i=1}^n (x_i\theta + \varepsilon_i) x_i}{\sum_{i=1}^n x_i^2} =
\mathsf{E}(\theta + \frac{\sum_{i=1}^n \varepsilon_i x_i}{\sum_{i=1}^n x_i^2}) = 
\theta + \frac{\sum_{i=1}^n \mathsf{E}(\varepsilon_i) x_i}{\sum_{i=1}^n x_i^2} = 
\theta + \frac{\sum_{i=1}^n 0 * x_i}{\sum_{i=1}^n x_i^2} = \theta$

**Часть 2.**\
Имеем $x \in \mathbb{R}$ &mdash; новый объект.\
$\mathsf{E}\widehat{y}(x) = 
\mathsf{E}(\widehat{\theta}*x) = \theta * x$

### Решение #4.

**Условие**\
$\widehat{\theta} = \frac{\sum_{i=1}^n Y_i x_i}{\sum_{i=1}^n x_i^2}$\
$\mathsf{E}\varepsilon_i = 0$\
$\mathsf{D}\varepsilon_i = \sigma^2$\
все $\varepsilon_i$ независимы

**Часть 1.**\
$\mathsf{D}\widehat{\theta} = \mathsf{D}\frac{\sum_{i=1}^n (x_i\theta + \varepsilon_i) x_i}{\sum_{i=1}^n x_i^2} =
\mathsf{D}(\theta + \frac{\sum_{i=1}^n \varepsilon_i x_i}{\sum_{i=1}^n x_i^2}) = 
\frac{\mathsf{D}(\sum_{i=1}^n \varepsilon_i x_i)}{(\sum_{i=1}^n x_i^2)^2} =^{\text(*)}$\
$= \frac{\sum_{i=1}^n \mathsf{D}(\varepsilon_i) x_i^2}{(\sum_{i=1}^n x_i^2)^2} = 
\frac{\sigma^2}{\sum_{i=1}^n x_i^2}$

*Примечание: переход $\text{(*)}$ совершён по независимости $\varepsilon_i$*

**Часть 2.**\
Имеем $x \in \mathbb{R}$ &mdash; новый объект.\
$\mathsf{D}\widehat{y}(x) = 
\mathsf{D}(\widehat{\theta}*x) = \sigma^2\frac{x^2}{\sum_{i=1}^n x_i^2}$

#### 2. Подготовка данных

Опишите по шагам процедуру предобработки данных, которая будет включать работу с выбросами, преобразования вещественных и категориальных признаков, возможно, что-то еще.

**Процедура предобработки данных:**

**1.** Сократим количество полей. Исходя из их зависимости или анализа графиков.

**2.** Отбросим выбросы. По каждому признаку обрежем по .01 и .99 квантилям.

**3.** Развернём категориальные признаки в бинарные признаки и уберём по одному из них, чтобы не допустить зависимости между признаками.

**4.** Применим некоторые функции к некоторым параметрам

Реализуйте данную процедуру.

**Приведём несколько способов продготовить данные.**

```python
import copy

def drop_outliers(data, real_features):
    return data[(np.abs(stats.zscore(data[real_features])) < 3).all(axis=1)]

# Лишь преобразуем категориальные признаки
def hypothesis0(data, encoder):
    categorial_features = ['season', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']
    real_features = ['temp', 'atemp', 'hum', 'windspeed']
    target_feature = 'cnt'
    d = copy.deepcopy(data)
    d = drop_outliers(d, real_features)
    train_cat = encoder.fit_transform(d[categorial_features])
    train = np.hstack([d[real_features], train_cat])
    return train, np.array(d[target_feature])

# Убрали зависимые признаки
def hypothesis1(data, encoder):
    categorial_features = ['season', 'holiday', 'weekday', 'weathersit']
    real_features = ['temp', 'hum', 'windspeed']
    target_feature = 'cnt'
    d = copy.deepcopy(data)
    d = drop_outliers(d, real_features)
    train_cat = encoder.fit_transform(d[categorial_features])
    train = np.hstack([d[real_features], train_cat])
    return train, np.array(d[target_feature])

# Убрали некоторые признаки, исходя из анализа графиков, без трансформации
def hypothesis2(data, encoder):
    categorial_features = ['season', 'weathersit', 'workingday']
    real_features = ['temp', 'windspeed']
    target_feature = 'cnt'
    d = copy.deepcopy(data)
    d = drop_outliers(d, real_features)
    train_cat = encoder.fit_transform(d[categorial_features])
    train = np.hstack([d[real_features], train_cat])
    return train, np.array(d[target_feature])

# Убрали признаки, исходя из анализа графиков, с кастомными функциями
def hypothesis3(data, encoder):
    categorial_features = ['season', 'weathersit', 'workingday']
    real_features = ['temp', 'windspeed']
    target_feature = 'cnt'
    d = copy.deepcopy(data)
    d = drop_outliers(d, real_features)
    d["temp"] = (d["temp"] - 0.5) ** 2
    d["windspeed"] = d["windspeed"] - 0.2
    train_cat = encoder.fit_transform(d[categorial_features])
    train = np.hstack([d[real_features], train_cat])
    return train, np.array(d[target_feature])
```

```python
prepare = hypothesis0 # Выбираем гипотезу по умлочанию
encoder = OneHotEncoder(drop='first', sparse=False)
input_a, target_a = prepare(train, encoder)
```

#### 3. Обучение модели и ее анализ

Обучите линейную регрессию из `sklearn` на полученных данных и посмотрите на полученные оценки коэффициентов.

```python
model = LinearRegression(fit_intercept=True)
model.fit(input_a, target_a)
model.coef_
```

Посчитайте ошибку предсказания на тестовой выборке по метрикам RMSE, MAE, MAPE.

*Замечание.* Если в тестовой части тоже есть "большие" выбросы, их стоит убрать. Наша модель не должна пытаться их предсказать, но в то же время выбросы вносят существенный вклад в значение метрики. 

```python
def mean_absolute_percentage_error(y_true, y_pred):
    return 100 * (np.abs((y_true - y_pred) / y_true)).mean()

def print_measurements(true_, preds_):
    print(f"RMSE: {metrics.mean_squared_error(true_, preds_) ** 0.5}")
    print(f"MAE: {metrics.mean_absolute_error(true_, preds_)}")
    print(f"MAPE: {mean_absolute_percentage_error(true_, preds_)}")

input_b, target_b = prepare(test, encoder)
test_preds = model.predict(input_b)
print_measurements(target_b, test_preds)
```

Посмотрим на обучающих данных:

```python
print_measurements(target_a, model.predict(input_a))
```

Попробуйте посчитать оценку коэффициентов с помощью формулы, полученной на занятии. В чем заключается проблема?

*Подсказка.* Подумайте о зависимостях между признаками.

**Ответ:** как было замечено и на паре - при зависимых признаках мы получаем около нулевой детерминант, откуда при поиске коэффициентов возникает большая, обратная ему(детерминанту) величина, которая делает коэф-ты слишком большими. Наша модель становится очень чуткой к ошибкам, т.е. если она ошибается, то делает это сильно(из-за чего все виды показателей ошибки (RMSE/MAE/MAPE) очень выросли).

```python
coefs = np.linalg.inv(input_a.T.dot(input_a)).dot(input_a.T).dot(target_a)
coefs
```

```python
best_preds = input_b.dot(coefs.T)
print_measurements(target_b, best_preds)
```

### Задача 3.

Как можно было увидеть ранее, оценки коэффициентов становятся слишком большими при наличии некоторых проблем с данными. Чтобы этого не происходило, часто выполняют процедуру *регуляризации* в модели. 

Пусть $X \in \mathbb{R}^{n \times d}$ &mdash; матрица признаков, $Y \in \mathbb{R}^{n}$ &mdash; вектор значений целевого признака, $\theta \in \mathbb{R}^{d}$ &mdash; неизвестный вектор параметров. В методе наименьших квадратов решается задача
$$\| Y - X\theta \|^2 \longrightarrow \min_\theta.$$
Регуляризация заключается в искусственном усечении значений вектора оценок коэффициентов путем добавления его нормы к оптимизируемому функционалу. Тем самым решается задача
$$\| Y - X\theta \|^2 + \lambda \| \theta \|^2 \longrightarrow \min_\theta,$$
где $\lambda > 0$ &mdash; гиперпараметр модели, то есть число, которое задается пользователем. Такая модель называется **Ridge-регрессией**.

**1.** Найдите оценку $\theta$ в модели ridge-регрессии. Сравните ее с оценкой коэффициентов в методе наименьших квадратов. Помогает ли она избавиться от проблемы обратимости матрицы в МНК, которая обсуждалась на занятии?

**Решение.**

Рассмотрим $F(\theta) = \|Y - X\theta\|^2 + \lambda\|\theta\|^2$.
Решим задачу $F \rightarrow \min\limits_{{\theta} \in \mathbb{R}}$.\
$F(\theta) = (Y - X\theta)^T(Y - X\theta) + \lambda\theta^T\theta =$\
$= Y^T Y - Y^T X \theta - \theta^T X^T Y + \theta^T X^T X \theta + \lambda\theta^T\theta =$\
$= Y^T Y - 2 Y^T X \theta + \theta^T X^T X \theta + \theta^T \lambda E \theta$
 
$\nabla F(\theta) = -2 X^T Y + 2 X^T X \theta + 2\lambda E\theta$\
$\nabla F(\theta) = 0 \iff X^T Y = (X^T X + \lambda E)\theta \iff \theta = (X^T X + \lambda E)^{-1} X^T Y$

Получаем оценку $\widehat{\theta} = (X^T X + \lambda E)^{-1} X^T Y$

**Проблема решена.** Даже если признаки зависимы и $\|X^T X\| \approx 0$, то однако матрица $X^T X + \lambda E$ будет обратима, и при обращении будет иметь не слишком большие коэффициенты.  

**2.** Выпишите формулу поиска оценки коэффициентов методом *градиентного спуска* и *стохастического градиентного спуска* для модели ridge-регрессии. Не забудьте указать, как получается батч.

**Решение.**
* **Градиентный спуск**\
Выберем $\theta_0$ - некоторое приближение искомой оценки $\widehat{\theta}$.\
Выберем также $\eta \in [10^{-4}, 10^{-1}]$.\
Рассмотрим последовательность $\theta_{t+1} = \theta_t - \eta*\nabla F(\theta_t)$, где $\nabla F(\theta) = -2 X^T Y + 2 X^T X \theta + 2\lambda \theta$\
Устремляя $t\rightarrow\inf$, получим $\theta_t\rightarrow\widehat{\theta}$, где $\widehat{\theta}$ - искомая оценка.

* **Стохастический градиентный спуск. Mini batch версия.**\
В этом случае последовательность будет выглядеть так:\
$\theta_{t+1} = \theta_t - \eta * ((X_I^T X_I + \lambda E)\theta_t - X_I^T Y_I)$,\
где\
$\quad k$ - гиперпараметр, размер пакета.\
$\quad I = [i_1, ..., i_k], i_j \in \{1,2,...,n\}$, выбрано случайно для каждого j, t.\
$\quad X_I, Y_I$ - соответствующие матрицы со строками с индексами из $I$

$\quad\quad$*Примечание: под $\eta$ занесли $\frac{2n}{k}$*

**3.** Многие модели машинного обучения требуют *стандартизацию признаков*. Существует множество различных способов стандартизации, из которых наиболее распространены два следующих.
1. <a target="_blank" href="https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html"><it>StandardScaler</it></a>. Пусть $x_{ij}$ &mdash; значение признака $j$ для объекта $i$. Обозначим $m_j$ &mdash; выборочное среднее значение признака $j$ (функция `np.mean`), а $s^2_j$ &mdash; выборочную дисперсию признака $j$ (функция `np.var`). Тогда стандартизацией является преобразование
$$\widetilde{x}_{ij} = \frac{x_{ij} - m_j}{s_j}.$$
2. <a target="_blank" href="https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html"><it>MinMaxScaler</it></a>. Пусть $x_{ij}$ &mdash; значение признака $j$ для объекта $i$. Обозначим $m_j$ и $M_j$ &mdash; минимальное и максимальное значения признака $j$ соответственно. Тогда стандартизацией является преобразование
$$\widetilde{x}_{ij} = \frac{x_{ij} - m_j}{M_j - m_j}.$$

Для метода наименьших квадратов стандартизацию признаков можно не производить, однако для модели ridge-регрессии стандартизация необходима. Поясните это. Можно представить, что признаки имеют некоторые различные единицы измерения.

**Решение.**
- Поскольку в ridge-регрессии мы минимизируем $\|Y - X\theta\|^2 + \lambda\|\theta\|^2$, то в случае различных единиц измерения у данных - справедливая оценка $\theta$ будет иметь примерно соответствующую разницу в порядках величин (рассмотрим $\theta_i$ и $\theta_j$: назовём их очень большим и очень маленьким коэффициентами). Но тогда 'минимизация' $\|\theta\|$ будет сведена к минимизации очень большого коэффициента, а не решению исходной задачи.

Для начала реализуйте класс для работы с моделью ridge-регрессии. Интерфейс этого класса соответствует стандарту библиотеки `sklearn`, который использует также множество других библиотек.

```python
from sklearn.preprocessing import StandardScaler
import copy
```

```python
class RidgeRegression:
    def __init__(self, lambd=0.1, fit_intercept=True):
        '''
        Инициализация модели.
        
        Аргументы:
        lambd -- параметр регуляризации.
        fit_intercept -- добавить ли константный признак.
        '''
        
        self.lambd = lambd
        self.fit_intercept = fit_intercept
        
    def change_input(self, X):
        if (self.fit_intercept):
            return np.hstack((X, np.ones((X.shape[0],1))))
        return X
            
        
    def fit(self, X, Y):
        '''
        Обучение модели ridge-регрессии.
        
        Аргументы:
        X -- numpy-матрица объект-признак размера (n, d);
        Y -- numpy-вектор значений целевого признака размера (n,).
        '''
        X = self.change_input(X)
        n, d = X.shape
        self.coefs = np.linalg.inv(X.T.dot(X) + self.lambd * np.identity(d)).dot(X.T).dot(Y)
        return self
    
    
    def predict(self, X):
        '''
        Вычисление предсказаний.
        
        Аргументы:
        X -- numpy-матрица объект-признак размера (n, d);
        
        Возвращает:
        predictions -- предсказания в виде numpy-вектора размера (n,).
        '''
        X = self.change_input(X)
        
        predictions = X.dot(self.coefs.T)
        
        return predictions
```

### Задача 5.

Предположим, что имеющиеся у нас данные по велопрокату &mdash; большие данные. Конечно же в реальности это не так. Однако, для простой учебной задачи это в самый раз, чтобы ничего ни у кого не подвисало.



Напомним формулу:\
$\theta_{t+1} = \theta_t - \eta * ((X_I^T X_I + \lambda E)\theta_t - X_I^T Y_I)$

```python
import random
class MiniBatchGradientDescent:
    def __init__(self, batch_size, step, lambd, init_coefs, fit_intercept=True):
        self.batch_size_ = batch_size
        self.step_ = step
        self.lambd_ = lambd
        self.coefs_ = init_coefs
        self.fit_intercept_ = fit_intercept
    
    def fit(self, X, Y, iterations = 1000):
        self.X = self.change_input_(X)
        self.Y = Y
        for i in range(iterations):
            self.move_()
        
    def predict(self, X):
        return self.change_input_(X).dot(self.coefs_.T)
    
    def move_(self):
        indices = self.generate_indices_(0, self.X.shape[0] - 1)
        X = self.X[indices]
        Y = self.Y[indices]
        n, d = X.shape
        self.coefs_ = self.coefs_.T - self.step_ * (((X.T @ X) + self.lambd_ * np.identity(d)) @ self.coefs_.T - X.T @ Y)
        
    def change_input_(self, X):
        if (self.fit_intercept_):
            return np.hstack((X, np.ones((X.shape[0],1))))
        return X
    
    def generate_indices_(self, lower_bound, upper_bound):
        return [random.randint(lower_bound, upper_bound) for i in range(self.batch_size_)]        
```
